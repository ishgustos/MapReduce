# Index inversé

```python
# -*- coding: utf-8 -*-
"""
Created on Sun Jan 15 23:58:14 2017

@author: User
"""

import json
import collections
import functools
import itertools

#from be1mr1 import MapReduce

def readData(filename):
    with open(filename, mode='r', encoding='utf-8') as file:
        for line in file:
            record = json.loads(line)
            yield(record)
            
#index inversé
def mapper(data):
    book = data[0]
    words = data[1].split()
    
    return [ (w,book) for w in words]

def reducer(data):
    return (data[0], list(set(data[1])))
    
def shuffle(data_mapped):
      
    #organiser les données retournées par map
    items = collections.defaultdict(list) #crée un dictionnaire vide
        
    for k,v in itertools.chain.from_iterable(data_mapped):     #itertools.chain.from_iterable() permet de fusionner des listes 
        items[k].append(v)  

    return items.items()
    
def invertedIndex(filename):
    
    data = readData(filename)
    data_mapped = map(mapper,data)
    data_shuffled = shuffle(data_mapped)
    index = map(reducer,data_shuffled)
    
    for (k,w) in index:
        print(k,w)
```
